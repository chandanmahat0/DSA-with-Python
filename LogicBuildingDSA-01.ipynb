#### 1. Checking 'm' is multiple of 'n'

```python
def is_multiple(n, m):
    return True if (n % m == 0) else False

n = 78
m = 4
print(is_multiple(n, m))
```

#### 2. Check whether a number is even

```python
def is_even(k):
    return True if (k & 1 == 0) else False

print("4 is a Even number:", is_even(4))
print("11 is a Even number:", is_even(11))
```

#### 3. Python function minmax(data) returning smallest and largest values without using built‑in min or max

```python
def minmax(data):
    smallest = data[0]
    largest = data[0]

    for value in data:
        if value < smallest:
            smallest = value
        if value > largest:
            largest = value

    return (smallest, largest)

data = (1,2,3,4,5,6)
print(minmax(data))
```

```python
def minmax(data):
    smallest = [x for x in data if all(x <= y for y in data)][0]
    largest = [x for x in data if all(x >= y for y in data)][0]
    return (smallest, largest)

numbers = "RamChandra"
print(minmax(numbers))
```

```python
def maxmin(sequence):
    iterator = iter(sequence)
    smallest = largest = next(iterator)

    for value in iterator:
        if value < smallest:
            smallest = value
        elif value > largest:
            largest = value

    return smallest, largest

seq = 'Chandan Mahato'
print(maxmin(seq))
```

```python
def minmax(data):
    iterator = iter(data)
    first = next(iterator)
    smallest = first
    largest = first

    for number in iterator:
        if number < smallest:
            smallest = number
        if number > largest:
            largest = number

    return (smallest, largest)

test_cases = [
    [3, 1, 4, 1, 5, 9, 2, 6],
    [-10, 0, 10, -20, 20],
    [42],
    [7, 7, 7, 7],
    (100, 50, 75, 25, 99),
]

for data in test_cases:
    result = minmax(data)
    print(f"Data:   {data}")
    print(f"Result: min = {result[0]}, max = {result[1]}")
    print()
```

#### 4. Function returning the sum of squares of positive integers smaller than n

```python
def sumSquareTerms(term):
    if term > 0:
        return sum(value ** 2 for value in range(term))
    else:
        return "Give input of positive integers"

num = 100
print(sumSquareTerms(num))
```

```python
def addSquareTerms(n):
    if n > 0:
        return ((n + 1) // 2) * ((2 * n + 1) // 3)
    else:
        return "Give input of positive integers"

num = 50
print(addSquareTerms(num))
```
