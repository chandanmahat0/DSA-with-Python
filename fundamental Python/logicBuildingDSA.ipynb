#### 2. Checking 'k' is even or not, if yes return True else False. But, Your Function should use the multiplication, modulo and division Operators.

```python
def is_even(k):
  # Using Bitwise '&' Operator, if number is even then '&' operator return 0 else it returns 1.
  return ( True if (k & 1 == 0) else False ) 
print("4 is a Even number:",is_even(4))
print("11 is a Even number:",is_even(11))
```

#### 3. write a program of python function minmax(data) that takes a sequences of one or more numbers, and return the smallest and the largest numbers int the form of a tuple of length two do not use the built in function min or max during program implementation.

```python
def minmax(data):
  smallest = data[0]
  largest = data[0]
  for value in data:
    if value < smallest:
      smallest = value
    if value > largest:
      largest = value
  return (smallest, largest)

data = (1,2,3,4,5,6)
print(minmax(data))
type(minmax(data)) # Type of Output i.e. Tuple.
```

```python
def maxmin(sequence):
  iterator = sequence.__iter__() # iter(sequence)
  smallest = largest = next(iterator)
  for value in iterator:
    if value < smallest:
      smallest = value
    elif value > largest:
      largest = value
  return smallest, largest

seq = 'Chandan Mahato'
print(maxmin(seq))
type(maxmin(seq))
```
