#### 1. Sum of the squares of all the odd positive numbers smaller than 'n'.

```python
def sumOddSquares(term):
  sum = 0
  for value in range (1, term, 2):
    sum += value ** 2
  return sum

term = 5
print("Sum of square of first odd integers is: ", sumOddSquares(term))
```

```python
def add_odd_square_num(x):
  return sum(i ** 2 for i in range (1, x) if (i % 2 == 1))

no_term = 5
print(add_odd_square_num(no_term)) 
```

```python
# This is only applicable when of 'n' first odd number is given.
def formula_sum_odd_square(n):
  return (n * (2*n - 1) * (2*n + 1) // 3)

term = 2
print(formula_sum_odd_square(term))
```

#### 2. Python's random module includes a function choice(data) that returns a random element from a non-empty sequence. The random module includes a more basic function, that return a random choice from the given range. Using only the randrange fuction, implement your own version of the choice function.

```python
import random
def my_choice(data):
  index = random.randrange(len(data)) #Randrange return index of random selection.
  return data[index]

data = [4, 5, 2, 8, 2]
print(my_choice(data))
```
