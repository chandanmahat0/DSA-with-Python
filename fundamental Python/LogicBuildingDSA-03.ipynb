##### 1. Write a pseudo code description of a function that reberses a list of  n integers, so that the numbers are listed in the opposite order than they were before, and compare this method to an equivalent Python function for doing the same thing.

```python
# Method 1: Using slicing.
def dataReverse(data):
  return data[::-1] 

list_1 = [3, 6, 9, 2, 90, 100]
print(dataReverse(list_1))
```

```python
# Method 2: Using For - Loop.
def data_Reverse(data):
  new_data = []
  for i in range (1, len(data) + 1):
    new_data.append(data[-i])
  data = new_data
  return data

list_2 = [4, 6, 3, 99, 34, 22, 1]
print(data_Reverse(list_2))
```

```python
# Method 3: Using Reverse Function.
data = [1, 2, 3, 4, 5, 6]
data.reverse()
print(data)
```

##### 2. Write a short Python function that takes a sequence of integer values and deremines if their is a distinct pair of numbers in the sequence whose product is odd.

```python
def has_odd_product_pair(seq):
    odd_count = 0
    
    for num in seq:
        if num % 2 != 0:   # check if number is odd
            odd_count += 1
        
        if odd_count >= 2:
            return True
    
    return False
data = [1, 2, 3, 4, 5, 6]
print(has_odd_product_pair(data))
```

```python
def has_odd_product_pair(seq):
    return sum(num % 2 for num in seq) >= 2

data = [1, 2, 3, 4, 5, 6]
print(has_odd_product_pair(data))
```
